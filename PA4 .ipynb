```python
#import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file and refer to it as 'board'
file_path = "board2.xlsx"  # Update with the actual path to the file
df = pd.read_excel("board2.xlsx")
df
```

## A.
Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as
Instrumentation and hometown Luzon

```python
# 1a. Create "Instru" DataFrame
# Track is "Instrumentation", Hometown is "Luzon", Electronics > 70
instru_df = df[(df['Track'] == 'Instrumentation') & (df['Hometown'] == 'Luzon') & (df['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']]
instru_df['Track'] = 'Instrumentation'
instru_df['Hometown'] = 'Luzon'

# Display the Instru DataFrame
instru_df
```

## B.
Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is
constant as Mindanao and gender Female

```python

# 1b. Create "Mindy" DataFrame
# Hometown is "Mindanao", Gender is "Female", Average >= 55
df['Average'] = (df['Math'] + df['GEAS'] + df['Electronics']) / 3
mindy_df = df[(df['Hometown'] == 'Mindanao') & (df['Gender'] == 'Female') & (df['Average'] >= 55)][['Name', 'Track', 'Electronics', 'Average']]
mindy_df['Gender'] = 'Female'
mindy_df['Hometown'] = 'Mindanao'

# Display the Mindy DataFrame
mindy_df
```

## 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?                                                                                                                                         


```python

# Visualize how Track contributes to the average score
plt.figure(figsize=(8,6))
df.groupby('Track')['Average'].mean().plot(kind='bar')
plt.title('Average Grade by Track')
plt.ylabel('Average Grade')
plt.show()
```

```python
# Visualize how Gender contributes to the average score
plt.figure(figsize=(8,6))
df.groupby('Gender')['Average'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Grade by Gender')
plt.ylabel('Average Grade')
plt.xlabel('Gender')
plt.show()
```

```python
# Visualize how Hometown contributes to the average score
plt.figure(figsize=(8,6))
df.groupby('Hometown')['Average'].mean().plot(kind='bar', color='lightgreen')
plt.title('Average Grade by Hometown')
plt.ylabel('Average Grade')
plt.xlabel('Hometown')
plt.show()
```
